```python
import pandas as pd
import numpy as np
```

```python
def convert_months(month):
    
    months = {}
    months['Abr'] = '04'
    months['Ago'] = '08'
    months['Dez'] = '12'
    months['Fev'] = '02'
    months['Jan'] = '01'
    months['Jul'] = '07'
    months['Jun'] = '06'
    months['Mai'] = '05'
    months['Mar'] = '03'
    months['Nov'] = '11'
    months['Out'] = '10'
    months['Set'] = '09'
    
    return months[month]
```

```python
df = pd.read_csv('./mortality_data/Dados_SICO_2020-04-04.csv')
```

```python
df['aux'] = df['Data (mm-dd)'].str[:3]
df['month'] = df.aux.apply(lambda x: convert_months(x))
```
